```python
#array operations
A=np.array([12,56,78,3])
B=np.array([90,45,78,1])
C=A-B        #a list will throw an error while array iterates themselves
print(C)
print(A-B)
```

```python
#one on one multiplication
A=np.array([45,67,21,8]).reshape(2,2)
B=np.array([34,67,2,7]).reshape(2,2)
print(A)
print("---------------------------")
print(B)
print("---------------------------")
C=A*B
print(C)
```

```python
#matrix multiplication
A=np.array([45,67,21,8]).reshape(2,2)
B=np.array([34,67,2,7]).reshape(2,2)
print(A)
print("---------------------------")
print(B)
print("---------------------------")
C=A@B
print(C)
```

```python
#matrix multiplication
A=np.array([45,67,21,8]).reshape(2,2)
B=np.array([34,67,2,7]).reshape(2,2)
print(A)
print("---------------------------")
print(B)
print("---------------------------")
C=A .dot(B)
print(C)
```

```python
#max elements ammong each row and column
A=np.array([45,67,21,8,8,90,78,45]).reshape(4,2)
print(A)
print("-----------------------")
print(A.max(axis=0))           #axis=0 -column
print("-----------------------")
print(A.max(axis=1))           #axis=1 -row
```

```python
#min elements among each row and column
A=np.array([45,67,21,8,8,90,78,45]).reshape(4,2)
print(A)
print("-----------------------")
print(A.min(axis=0))           #axis=0 -column
print("-----------------------")
print(A.min(axis=1))           #axis=1 -row
```

```python
# joining arrays vertically
A=np.array([45,67,21,8]).reshape(2,2)
B=np.array([34,67,2,7]).reshape(2,2)
print(A)
print("---------------------------")
print(B)
print("---------------------------")
print("After vertically stacking the arrays")
print(np.vstack((A,B)))
```

```python
# joining arrays horizontally
A=np.array([45,67,21,8]).reshape(2,2)
B=np.array([34,67,2,7]).reshape(2,2)
print(A)
print("---------------------------")
print(B)
print("---------------------------")
print("After horizontally stacking the arrays")
print(np.hstack((A,B)))
```

```python
# joining arrays
A=np.array([45,67,21,8]).reshape(2,2)
B=np.array([34,67,2,7]).reshape(2,2)
print(A)
print("---------------------------")
print(B)
print("---------------------------")
print("After  stacking the arrays")
print(np.stack((A,B),axis=0))
```

```python
# joining arrays
A=np.array([45,67,21,8]).reshape(2,2)
B=np.array([34,67,2,7]).reshape(2,2)
print(A)
print("---------------------------")
print(B)
print("---------------------------")
print("After  stacking the arrays")
print(np.stack((A,B),axis=1))
```

```python
#dimentionality stack
A=np.arange(30).reshape(2,3,5)
print(A)
print("After dstack")
print(np.dstack(A))#no.of rows -->no.of groups
#no.of columns in every group -->no.of rows in every group
#            g   r  c
#original   2    3  5
#changed    5   2   3
#changed    c   g   r
```
